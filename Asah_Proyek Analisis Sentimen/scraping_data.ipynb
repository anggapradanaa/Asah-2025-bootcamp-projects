# **Import Library**

```python
import pandas as pd
from google_play_scraper import reviews, Sort
import os
```

```python
def scrape_gojek_reviews_simple(target_total=12000):
    app_id = 'com.gojek.app'
    print(f"Mengambil {target_total} ulasan dari aplikasi Gojek")
    
    all_reviews = []
    continuation_token = None

    while len(all_reviews) < target_total:
        try:
            result, continuation_token = reviews(
                app_id,
                lang='id',
                country='id',
                sort=Sort.NEWEST,
                count=200,
                continuation_token=continuation_token
            )
            
            # Tambahkan semua review yang didapat
            for review in result:
                if len(all_reviews) < target_total:
                    all_reviews.append(review)
                else:
                    break

            print(f"Progress: {len(all_reviews)}/{target_total} ulasan terkumpul")

            # Cek apakah target tercapai
            if len(all_reviews) >= target_total:
                print(f"\nTarget {target_total} ulasan tercapai.")
                break

            if continuation_token is None:
                print(f"\nTidak ada ulasan lagi. Total terkumpul: {len(all_reviews)}")
                break

        except Exception as e:
            print(f"\nTerjadi error: {e}")
            break

    # Ekstrak data ke DataFrame
    data_ulasan = []
    for review in all_reviews:
        data_ulasan.append({
            'userName': review['userName'],
            'score': review['score'],
            'content': review['content'],
            'thumbsUpCount': review['thumbsUpCount'],
            'reviewCreatedVersion': review['reviewCreatedVersion'],
            'at': review['at'],
            'replyContent': review['replyContent'],
            'repliedAt': review['repliedAt']
        })

    df = pd.DataFrame(data_ulasan)
    print(f"\nTotal ulasan terkumpul: {len(df)}")
    print("\nDistribusi rating:")
    print(df['score'].value_counts().sort_index())
    return df


if __name__ == "__main__":
    df_reviews = scrape_gojek_reviews_simple(target_total=12000)

    # Simpan hasil
    folder_path = r'D:\Perkuliahan\Asah led by Dicoding\Asah_Proyek Analisis Sentimen'
    os.makedirs(folder_path, exist_ok=True)
    filename = 'gojek_reviews_v2.csv'
    filepath = os.path.join(folder_path, filename)

    df_reviews.to_csv(filepath, index=False, encoding='utf-8-sig')
    print(f"\nData berhasil disimpan ke: {filepath}")
```
